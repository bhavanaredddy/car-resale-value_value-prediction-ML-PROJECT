# EDA FOR BIVARIATE ANALYSIS
*bivariate analysis involves exploring the relationship between two variables in a dataset. The primary goal is to understand the patterns, associations, and dependencies between the two variables

*Know the variables you want to analyze and their data types (e.g., numerical or categorical).

*Determine the research question or hypothesis you want to investigate.

*Handle missing data: Check for missing values and decide how to handle them (e.g., impute or remove).

*Remove outliers that could significantly affect the analysis.

*Use scatter plots for two numerical variables to visualize their relationship. Each data point represents a pair of values for the two variables.

*For a categorical variable and a numerical variable, you can use box plots, violin plots, or bar plots to visualize how the numerical variable varies across different categories.

*For two categorical variables, create a contingency table and use stacked bar plots or mosaic plots to visualize the relationship between the two variables.

*Calculate the correlation coefficient (e.g., Pearson's correlation for numerical variables or Cramer's V for categorical variables) to quantify the strength and direction of the relationship.

*Generate a correlation matrix or heatmap to visualize the correlations between multiple pairs of numerical variables.

*Based on the analysis, draw conclusions and generate insights about the relationship between the two variables.

*Be cautious not to make causal claims unless you have a well-designed experiment.

# Removing unwanted labels
    we can remove the owner column

```python
df = df.drop('Owner',axis=1)
```

# Label encoding values

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df['Car_Name'] = le.fit_transform(df['Car_Name'])
df['Year'] = le.fit_transform(df['Year'])
df['Selling_Price'] = le.fit_transform(df['Selling_Price'])
df['Present_Price'] = le.fit_transform(df['Present_Price'])
df['Kms_Driven'] = le.fit_transform(df['Kms_Driven'])
df['Fuel_Type'] = le.fit_transform(df['Fuel_Type'])
df['Seller_Type'] = le.fit_transform(df['Seller_Type'])
df['Transmission'] = le.fit_transform(df['Transmission'])
```
